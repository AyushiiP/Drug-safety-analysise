```python
# Define the function to categorize blood pressure, with error handling
def categorize_bp(bp):
    try:
        sys, dia = map(int, bp.split('/'))
        if sys >= 140 or dia >= 90:
            return 'High'
        elif sys < 120 and dia < 80:
            return 'Normal'
        else:
            return 'Elevated'
    except ValueError:
        return 'Unknown'  # Assign 'Unknown' if the BP value is non-numeric or improperly formatted
```

```python
# Apply the function to the 'BP' column to create a new 'BP Category' column
df['BP Category'] = df['BP'].apply(categorize_bp)
```

```python
# Plot blood pressure categories
plt.figure(figsize=(5,3))
sns.countplot(x='BP', data=data)
plt.title('Blood Pressure Categories Distribution')
plt.xlabel('Blood Pressure Category')
plt.ylabel('Count')
plt.show()
```
